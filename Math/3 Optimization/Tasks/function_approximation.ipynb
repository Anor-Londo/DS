```python
def gen_coefs_matrix(points):
    res_matrix = []
    for point in points:
        pol = []
        for i in range(len(points)):
            coef_value = point ** i 
            pol.append(coef_value)
        res_matrix.append(pol)
    return np.array(res_matrix)
```

```python
def gen_free_vector(points):
    res_vector = [func(point) for point in points]
    return np.array(res_vector)[:, np.newaxis]
```
