```python
import math
import numpy as np
from scipy.optimize import minimize
from matplotlib import pyplot as plt
```

```python
trace = []
def func(x):
    res = np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
    trace.append((x, res))
    return res
```

```python
X = np.arange(-1, 30)
Y = [func(x) for x in X]
```

```python
plt.plot(X, Y)
start, color = 2, 'red'
trace = []
solution = minimize(func, start, method='BFGS')
x, y = solution.x[0], solution.fun
print(f'{x:.2f}, {y:.2f}')
plt.plot(*zip(*trace), '.-', c=color, label=f'start {start:.1f}')
plt.plot(*trace[0], 'o', c=color)
#plt.text(x, -23, f'{x:.3f}', c=color, ha='center')
#plt.plot([x, x], [1, y], '--', c=color)
plt.legend()
plt.show()
```
